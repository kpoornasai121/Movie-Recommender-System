```python
import ast
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
convert(movies.iloc[374].genres)
```

```python
movies['genres'].apply(convert)
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter!=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['cast'].apply(convert3)
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
  l=[]
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head(1)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.get_feature_names_out()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stemming(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
%%writefile app.py
import streamlit as st
import pickle
import requests
import simplejson as json
movies=pickle.load(open('movies.pkl','rb'))
similarity=pickle.load(open('similarity.pkl','rb'))

def fetch_poster(movie_id):
  url='https://api.themoviedb.org/3/movie/{}?api_key=2736a08daef7a534d3cf2d8c371e0427&language=en-US'.format(movie_id)
  data=requests.get(url)
  data=data.json()
  poster_path=data['poster_path']
  full_path="https://image.tmdb.org/t/p/w500"+data['poster_path']
  return full_path

def recommend(movie):
  index=movies[movies['title']==movie].index[0]
  distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
  recommended_movies=[]
  recommended_movies_posters=[]
  for i in distances[1:11]:
    movie_id=movies.iloc[i[0]].movie_id
    recommended_movies.append(movies.iloc[i[0]].title)
    recommended_movies_posters.append(fetch_poster(movie_id))
  return recommended_movies,recommended_movies_posters


st.title('Movie Recommender System')
movie_list=movies['title'].values
selected_movie=st.selectbox("Type or select a movie from the dropdown",movie_list)

if st.button('Show Recommendation'):
  recommended_movies,recommended_movies_posters=recommend(selected_movie)
  c1,c2,c3,c4,c5,c6,c7,c8,c9,c10=st.tabs(['Movie 1','Movie 2','Movie 3','Movie 4','Movie 5','Movie 6','Movie 7','Movie 8','Movie 9','Movie 10'])   #st.columns(10)
  with c1:
    st.subheader(recommended_movies[0])
    st.image(recommended_movies_posters[0],width=220)
  with c2:
    st.subheader(recommended_movies[1])
    st.image(recommended_movies_posters[1],width=220)
  with c3:
    st.subheader(recommended_movies[2])
    st.image(recommended_movies_posters[2],width=220)
  with c4:
    st.subheader(recommended_movies[3])
    st.image(recommended_movies_posters[3],width=220)
  with c5:
    st.subheader(recommended_movies[4])
    st.image(recommended_movies_posters[4],width=220)
  with c6:
    st.subheader(recommended_movies[5])
    st.image(recommended_movies_posters[5],width=220)
  with c7:
    st.subheader(recommended_movies[6])
    st.image(recommended_movies_posters[6],width=220)
  with c8:
    st.subheader(recommended_movies[7])
    st.image(recommended_movies_posters[7],width=220)
  with c9:
    st.subheader(recommended_movies[8])
    st.image(recommended_movies_posters[8],width=220)
  with c10:
    st.subheader(recommended_movies[9])
    st.image(recommended_movies_posters[9],width=220)
```
